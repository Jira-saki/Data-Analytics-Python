
## How many people have credit card expire in 2022

```python
df.columns
```

```python
df['CC Exp Date']
```

```python
# Split data 03/22 to be 2 numbers

def func():
    """
     split data into list[]
     count+1 on index[1] if value == 22
    """
    count=0
    for date in df['CC Exp Date']:
        if date.split('/')[1] == '22':
            count=count+1
    print(count)
        
```

```python
func()
```

### Verify Top 5 most using mail providers(e.g. gmail.com, hotmail.com etc...) 

```python
# make a list[] of email split column
list1 = []
for email in df['Email']:
    list1.append(email.split('@')[1])
```

```python
# create temp dataframe to count emails
df['temp']= list1
df.head(1)
```

```python
df['temp'].value_counts().head(5)
```
