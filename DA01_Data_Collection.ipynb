```
# list all tables
cursor = connection.cursor()
cursor.execute("show tables;")
tables = cursor.fetchall()
cursor.close()
print(tables)
```

```
# with statement instaed of cursor.close()

with connection.cursor() as cursor:
  # Read a single record
  sql = "SELECT * FROM online_retail;"
  cursor.execute(sql)
  result = cursor.fetchall()

print("number of rows: ", len(result))
```

# 2. Get data from REST API

### Using package `requests` for  REST API HTTP GET
 

*   get API currency rate from UK Pound to Thai Baht
*   Convert to dataframe
*   Join the data



```
import requests
```



## Requests library
Docs of package [requests](https://requests.readthedocs.io/en/master/)

```
url = "https://de-training-2020-7au6fmnprq-de.a.run.app/currency_gbp/all"
response = requests.get(url)
```

```
type(response.text)
```

## JSON loading

import `json` (built-in package)
  
Convert JSON string to python dictionary.



```
import json
```

```
? json
```

```
result_conversion_rate = json.loads(response.text)
result_conversion_rate
```

```
print(type(result_conversion_rate))
assert isinstance(result_conversion_rate, dict)
```

###Convert to result_conversion_rate(dict) to Pandas dataframe


```
# key is Time data in index column
# use 'from_dict()' for key of Dictionary
conversion_rate = pd.DataFrame.from_dict(result_conversion_rate)
```

```
conversion_rate[:3]
```

Naming index column to "date"

```
conversion_rate = conversion_rate.reset_index().rename(columns={"index":"date"})
conversion_rate[:3]
```

# Join the data

we are going to combine 2 dataframes(retail DF + Conversion_rate DF+

* Use *InvoiceDate* column of retail DF 

* Use *date* column in conversion_rate DF


* Data in *InvoiceDate* is datetime while
data in  *date*  is timestamp (all on 00:00:00) 

```
retail.info()
```

```
# Copy column InvoiceDate to be InvoiceTimestamp for future use
retail['InvoiceTimestamp'] = retail['InvoiceDate']
retail
```

```
# Convert InvoiceDate in retail only for date without time
# Convert date in conversion_rate only for date
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate']).dt.date
conversion_rate['date'] = pd.to_datetime(conversion_rate['date']).dt.date

retail.head()
```

```
# Merge 2 dataframe 
final_df = retail.merge(conversion_rate, how="left", left_on="InvoiceDate", right_on="date")
final_df
```

```
# Add column 'THBPrice' (column UnitPrice * Rate)

# final_df['THBPrice'] = final_df['UnitPrice'] * final_df['Rate']
final_df['THBPrice'] = final_df.apply(lambda x: x['UnitPrice'] * x['Rate'], axis=1)  
final_df
```

## Save to CSV

```
#save "to csv" file
final_df.to_csv("output.csv", index=False)
```
